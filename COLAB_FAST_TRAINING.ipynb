## Step 6: Download Trained Models (after training completes)

```python
# Create a zip file of trained models to download
import shutil
import glob

source_dir = '/content/AI-Talent-Intelligence-Platform/django_pg_backend/core/apps/analytics/management/trained_models'
output_file = '/content/trained_models_backup.zip'

shutil.make_archive(output_file.replace('.zip', ''), 'zip', source_dir)
print(f"Models backed up to: {output_file}")

# List trained models
print("\nTrained models contents:")
for f in glob.glob(f"{source_dir}/*"):
    print(f"  {os.path.basename(f)}")
```
