## Step 5: Download Trained Models (after training completes)

```python
import shutil
import glob

# Create a zip file of trained models
source_dir = '/content/drive/My Drive/AI-Talent-Intelligence-Platform/django_pg_backend/core/apps/analytics/management/trained_models'
output_file = '/content/drive/My Drive/AI-Talent-Intelligence-Platform/trained_models_backup.zip'

shutil.make_archive(output_file.replace('.zip', ''), 'zip', source_dir)
print(f"Models backed up to: {output_file}")

# Also list what's in the trained_models folder
print("\nTrained models contents:")
for f in glob.glob(f"{source_dir}/*"):
    print(f"  {os.path.basename(f)}")
```
